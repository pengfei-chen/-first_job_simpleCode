###  计算相似度

```python
from sklearn.metrics.pairwise import cosine_similarity
i = [1,0,0,0]
j = [1,0.5,0.5,0]
cosine_similarity([i,j])
```

```python
from scipy.stats import pearsonr

i = [1,0,0,0]
j = [1,0.5,0.5,0]
pearsonr(i,j)
```
