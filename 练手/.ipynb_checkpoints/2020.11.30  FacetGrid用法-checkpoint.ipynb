```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
tips
```

```python
grid = sns.FacetGrid(tips, row="sex", col="time", hue="smoker",margin_titles=True)  # hue表示颜色
grid;
```

```python
grid.map(plt.scatter, "total_bill","tip");
grid.add_legend();
```

###  FacetGrid

在设定条件下，查看想要查看变量之间的关系！

比如上图：在time、sex 两种不同组合的情况下，查看变量total_bill，和 tip 的关系
