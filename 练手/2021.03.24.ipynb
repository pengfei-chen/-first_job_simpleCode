```python
# 创建停用词list
def stopwordslist(filepath):
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]
    return stopwords
```

```python
# 对句子进行分词
def seg_sentence_segment(sentence):
    stopwords = stopwordslist('stopwords.txt')  # 这里加载停用词的路径
    seg = HanLP.segment(sentence)
    outstr = ''
    for word in seg:
        if word.word not in stopwords:
            if word != '\t':
                outstr += word.word
                outstr += " "
    return outstr
```

```python
result = ''
with open('./data/美集会纪字[2018]038号  集团8月经营分析会纪要.txt','r', encoding='utf-8') as f:
    for line in f:
        line_seg = seg_sentence_segment(line)
        result += line_seg
result
```

```python
result = result.replace('\n','')
result = result.replace('\u3000','')
result = result.replace('\ufeff[','')
result = result.split(" ")
result = list(filter(lambda x: x!='', result))
result
```

```python
# 读取文件夹内的所有文件
path = './data/'
file_dir = os.listdir(path)
use = ''
for file in file_dir:
    file_name = path + file
    with open(file_name,'r', encoding='utf-8') as f:
        for line in f:
            line_seg = seg_sentence_segment(line)  # 这里的返回值是字符串
            use += line_seg
len(use)
```

```python
use = use.replace('\n','')
use = use.replace('\u3000','')
use = use.replace('\ufeff[','')
use = use.split(" ")
use = list(filter(lambda x: x!='', use))
use[:2000]
```

### 构造字典，统计其出现的次数，并从大到小排序

```python
words_index = set(use)
res = []
for i in words_index:
    if len(i) != 1:
        res.append(i)
res
```

```python
#创建一单词为key，频率为值的字典
counts_dict = {index:use.count(index) for index in res}
counts_dict
```

```python
#其中利用lambda表达式，以字典中的值为排序的参数
result = sorted(counts_dict.items(),key=lambda x: x[1],reverse=True)
result = pd.DataFrame(result,columns=['word','times'])
result
```

```python
result.to_csv('./result_0324.csv',index=False, encoding='gb18030')
```
